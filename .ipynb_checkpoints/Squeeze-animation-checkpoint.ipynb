```python
from IPython import display
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter

# definition of the order of electrodes
ELECTRODES = {"C0":36,"HV1":35}
for i in range(2,21):
    ELECTRODES[f"C{i-1}"]=36 - i
ELECTRODES["HV2"]=36 - 21
for i in range(22,36):
    ELECTRODES[f"P{i-21}"]=36 - i
ELECTRODES["HV3"] = 0

V_vis = [0]*len(ELECTRODES)

nested_trap_new = {
    "inlet": ["C2"],
    "floor": ["C3", "C4", "C5", "C6", "C7", "C8", "C9", "C10", "C11", "C12", "C13", "C14", "C15", "C16", "C17", "C18", "C19", "HV2", "P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10", "P11", "P12"],
    "outlet": ["P13"],
    "predump": ["C3", "C4", "C5", "C6", "C7", "C8", "C9", "C10", "C11", "C12", "C13", "C14", "C15", "C16", "C17", "C18", "C19", "HV2", "P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10"],
    "2E_MCP": ["C3", "C4", "C5", "C6", "C7", "C8", "C9", "C10", "C11", "C12", "C13", "C14", "C15", "C16", "C17", "C18", "C19", "HV2", "P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10"],
    "4E_MCP": ["C3", "C4", "C5", "C6", "C7", "C8", "C9", "C10", "C11", "C12", "C13", "C14", "C15", "C16", "C17", "C18", "C19", "HV2", "P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8"],
    "6E_MCP": ["C3", "C4", "C5", "C6", "C7", "C8", "C9", "C10", "C11", "C12", "C13", "C14", "C15", "C16", "C17", "C18", "C19", "HV2", "P1", "P2", "P3", "P4", "P5", "P6"],
    "8E_MCP": ["C3", "C4", "C5", "C6", "C7", "C8", "C9", "C10", "C11", "C12", "C13", "C14", "C15", "C16", "C17", "C18", "C19", "HV2", "P1", "P2", "P3", "P4"],
    "Half_MCP": ["C3", "C4", "C5", "C6", "C7", "C8", "C9", "C10", "C11", "C12", "C13", "C14", "C15", "C16", "C17"],
    "2E_ELENA_start": ["P12", "P11", "P10", "P9", "P8", "P7", "P6", "P5", "P4", "P3", "P2", "P1", "HV2", "C19", "C18", "C17", "C16", "C15", "C14", "C13", "C12", "C11", "C10", "C9", "C8", "C7", "C6", "C5", "C3"],
    "2E_ELENA_end": ["C6", "C7", "C8", "C9", "C10", "C11", "C12", "C13", "C14", "C15", "C16", "C17", "C18", "C19", "HV2", "P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10", "P11", "P12", "P13"],
    "2E_ELENA_MRTOF_end": ["C6", "C7", "C8", "C9", "C10", "C11", "C12", "C13", "C14", "C15", "C16", "C17", "C18", "C19", "HV2", "P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10", "P11", "P12"],
    "2E_MCP_SR3": ["P12","P11"],
    "2E_ELENA_SR4": ["C4"] }

nested_trap = {
    "inlet": ["C2"],
    "floor": ["C3", "C4", "C5", "C6", "C7", "C8", "C9", "C10", "C11", "C12", "C13", "C14", "C15", "C16", "C17", "C18", "C19", "HV2", "P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10", "P11", "P12"],
    "outlet": ["P13"],
    "predump": ["C3", "C4", "C5", "C6", "C7", "C8", "C9", "C10", "C11", "C12", "C13", "C14", "C15", "C16", "C17", "C18", "C19", "HV2", "P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10"],
    "2E_MCP": ["P10","P9","P8","P7","P6","P5","P4","P3","P2","P1","HV2","C19","C18","C17","C16","C15","C14","C13","C12","C11","C10","C9","C8","C7","C6","C5","C4","C3"],
    "4E_MCP": ["P8","P7","P6","P5","P4","P3","P2","P1","HV2","C19","C18","C17","C16","C15","C14","C13","C12","C11","C10","C9","C8","C7","C6","C5","C4","C3"],
    "6E_MCP": ["P6","P5","P4","P3","P2","P1","HV2","C19","C18","C17","C16","C15","C14","C13","C12","C11","C10","C9","C8","C7","C6","C5","C4","C3"],
    "8E_MCP": ["P4","P3","P2","P1","HV2","C19","C18","C17","C16","C15","C14","C13","C12","C11","C10","C9","C8","C7","C6","C5","C4","C3"],
    "Half_MCP": ["C17","C16","C15","C14","C13","C12","C11","C10","C9","C8","C7","C6","C5","C4","C3"],
    "2E_ELENA_start": ["C3", "C5", "C6", "C7", "C8", "C9", "C10", "C11", "C12", "C13", "C14", "C15", "C16", "C17", "C18", "C19", "HV2", "P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10", "P11", "P12"],
    "2E_ELENA_end": ["C6", "C7", "C8", "C9", "C10", "C11", "C12", "C13", "C14", "C15", "C16", "C17", "C18", "C19", "HV2", "P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10", "P11", "P12", "P13"],
    "2E_ELENA_MRTOF_end": ["C6", "C7", "C8", "C9", "C10", "C11", "C12", "C13", "C14", "C15", "C16", "C17", "C18", "C19", "HV2", "P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10", "P11", "P12"],
    "2E_MCP_SR3": ["P12","P11"],
    "2E_ELENA_SR4": ["C4"]  }

def ResetV(Vfloor=0,Vwall=0):
    for i in range(len(ELECTRODES)):
        V_vis[i] = Vfloor
    V_vis[ELECTRODES[nested_trap["inlet"][0]]] = Vwall
    V_vis[ELECTRODES[nested_trap["outlet"][0]]] = Vwall
    

def SetV(electrode,voltage):
    V_vis[ELECTRODES[electrode]] = voltage

def NiceSlowReshapeConfig(electrodes,Vstart, Vend, steps):
    V_animate = []
    # plt.ion()
    # fig = plt.figure()
    # ax = fig.add_subplot(111)
    
    iteration = 0
    # check if we are dividing by 0
    if steps == 0:
        steps = 1
    if Vend > 160:
        Vend = 160 # coerce
    # calculate voltage step size
    dV = (Vend - Vstart)/steps
    # define counters for electrodes
    start_electrode = 0
    stop_electrode = 1

    # get number of electrodes
    n_electrodes = len(electrodes)

    # create an array of voltages set
    V = [Vstart]*n_electrodes

    while abs(V[n_electrodes-1]-Vstart) < abs(Vend-Vstart):
        # loop over electrodes that we want to update
        for electrode in range(start_electrode,stop_electrode):
            # calculate new voltage for the electrodes 
            V[electrode] = V[electrode] + dV
            if abs(V[electrode]-Vstart) > abs(Vend-Vstart):
                V[electrode] = Vend
                if electrode == 0:
                    start_electrode = start_electrode+1
            # check if the value doesn"t extend the limit
            # # set new voltage for the electrode
            SetV(electrodes[electrode],V[electrode])
            # delay(10*ns)

        # # actually raise potential on all electrodes to set value    
        # for fastino in range(len(self.fastinos)):
        #     self.fastinos[fastino].update(0xFFFFFFFF)
        #     delay(10*ns)

        # update start_electrode
        # if abs(V[start_electrode]) >= abs(Vend):
        #     start_electrode = start_electrode+1
        # add another electrode if not all are already included
        if stop_electrode < n_electrodes:
            stop_electrode = stop_electrode + 1
            
        # print("#{}:".format(iteration))
        # print("  V={}".format(V))
        # print("  start_electrode={}".format(start_electrode))
        # print("  stop_electrode={}".format(stop_electrode))

        # delay(determined_delay)
        
        # # print([Vend-Vi for Vi in V])
        # ax.clear()
        # #ax.plot(V)
        # ax.stairs(V_vis)
        # ax.set_xticks(range(len(ELECTRODES)))
        # ax.set_xticklabels(list(ELECTRODES.keys()))
        # #plt.plot(V)
        # plt.xlabel("electrode")
        # plt.ylabel("voltage [V]")
        # plt.ylim(-200,200)
        # plt.xticks(rotation=90)
        # #plt.title("iteration={}".format(iteration))
        
        # display.clear_output(wait=True)
        # display.display(fig)
        iteration = iteration+1
        V_animate.append([i for i in V_vis])
    return V_animate

```

```python
Vwall = 160
Vfloor = -Vwall
steps = 100

electrodes = nested_trap_new["2E_MCP"]

ResetV(Vfloor,Vwall)
V_animate = NiceSlowReshapeConfig(electrodes,Vfloor,Vwall,steps)

fig,ax = plt.subplots()

def animate(i):
    ax.clear()
    ax.set_ylim(-200,200)
    ax.set_xticks(range(len(ELECTRODES)))
    names = list(ELECTRODES.keys())
    names.reverse()
    ax.set_xticklabels(names)
    ax.set_xlabel("electrode")
    ax.set_ylabel("voltage [V]")
    line = ax.stairs(V_animate[i])
    plt.xticks(rotation=90)
    return line, 
        
ani = FuncAnimation(fig, animate, interval=40, blit=True, repeat=True, frames=len(V_animate))    
ani.save("squeeze_MCP_new.gif", dpi=150, writer=PillowWriter(fps=25))
```
